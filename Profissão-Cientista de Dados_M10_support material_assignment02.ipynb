````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn

# Fixing random state for reproducibility
np.random.seed(2360873)
N = 100
````

````python
#carregar arquivo
df=pd.read_csv('previsão_renda_fil.csv')
df.drop('Unnamed: 0',axis=1, inplace=True)
df.head()
````

````python
# Selecionar apenas as colunas numericas
cols = ['posse_de_veiculo', 'posse_de_imovel', 'renda', 'sexo']
df_selected = df[cols]

# Plotar a matriz de dispersão
sns.pairplot(df_selected.drop('sexo',axis=1))
````

````python
ax = sns.pointplot(x="posse_de_veiculo", y="renda",
                   data=df_selected, dodge=True,  errorbar=('ci', 90))
````

````python
ax = sns.pointplot(x="posse_de_imovel", y="renda",
                   data=df_selected, dodge=True,  errorbar=('ci', 90))
````

````python
ax = sns.pointplot(x="posse_de_imovel", y="renda", hue = 'sexo',
                   data=df_selected, dodge=True,  errorbar=('ci', 95))

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)
````

````python
ax = sns.pointplot(x="posse_de_veiculo", y="renda", hue = 'sexo',
                   data=df_selected, dodge=True,  errorbar=('ci', 95))

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)
````

````python
# Plotar o mapa de calor da matriz de correlação
plt.figure(figsize=(20, 16))
sns.heatmap(df_selected.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlação')
````

As duas variaveis são boas preditoras
